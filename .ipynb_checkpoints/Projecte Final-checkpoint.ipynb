### 1) Nuevo fichero General de Entidades Ciudadanas de la ciudad de Barcelona (12/04/2023)
El Fichero General de Entidades Ciudadanas es uno de los instrumentos de relación del Ayuntamiento de Barcelona con las entidades y grupos de hecho de la ciudad. Su creación está prevista en las Normas reguladoras de la participación ciudadana (artículo 9) y está reglamentado mediante las Normas reguladoras del Fichero general de entidades ciudadanas (BOP 26/2/2004).

- Gerencia: Gerència de Drets de Ciutadania, Participació i Transparència  
- Departamento: Departament d'Associacionisme i Iniciatives Ciutadanes  
- Fuente: Comissionat de Participació Ciutadana i Associacionisme  
- Agenda 2030. ODS Principal: ODS 16: Paz, justícia e instituciones sólidas  
- Agenda 2030. ODS Colateral 1: ODS 17: Alianza para lograr los objetivos  
- Frecuencia actualización: Mensual  
- Información histórica: Sí

#### Explicación de las variables del fichero:  

01.**Tipologia_E_GF**: Tipologia de l'entitat. Indica si es tracta d'una entitat o grup de fet. Totes les entitats tenen aquest camp amb el valor "E"

13.**Nom_Districte**: Nom del districte  
14.**Codi_Barri**: Codi del barri  
15.**Nom_Barri**: Nom del barri  
16.**Url_Preferent**: URL preferent  
17.**Any_Constitucio**: Any de constitució de l'entitat  
18.**Sector_Territorial**: Territori en el qual l'entitat desenvolupa les seves activitats  
19.**Total_Socis**: Nombre total de socis de l'entitat  
20.**Telefon_Preferent**: Telèfon preferent de contacte  
21.**Correu_Preferent**: Adreça preferent de correu electrònic  
22.**Sector_Tematic_1**: Primer sector temàtic

```python
Entidades_sector_1_counts = df_entidades["Sector_Tematic_1"].value_counts()
Entidades_sector_2_counts = df_entidades["Sector_Tematic_2"].value_counts()
Entidades_sector_3_counts = df_entidades["Sector_Tematic_3"].value_counts()

# Combino las tres series en una sola, poniendo cero donde no tenga alguno de los temas de la serie
series = Entidades_sector_1_counts.add(Entidades_sector_2_counts, fill_value=0).add(Entidades_sector_3_counts, fill_value=0)

#la ordeno de mayor a menor
series_ordenadas = series.sort_values(ascending=False)
series_ordenadas
```

```python
# Obtengo una lista de todos los temas a partir del indice de la serie que he combinado
temas_unicos = series_ordenadas.index.tolist()

# Creo un DataFrame donde cada fila es un tema y hay una columna por sector. Relleno con los valores de las series y poniendo cero donde no hay valor
data = {
    "Sector Tematic 1": Entidades_sector_1_counts.reindex(temas_unicos, fill_value=0),
    "Sector Tematic 2": Entidades_sector_2_counts.reindex(temas_unicos, fill_value=0),
    "Sector Tematic 3": Entidades_sector_3_counts.reindex(temas_unicos, fill_value=0)
}

df = pd.DataFrame(data)
df
```

```python
# Creo el gráfico de barras apiladas
plt.figure(figsize=(16, 12))
df.plot(kind="bar", stacked=True)

plt.xlabel("Temas")
plt.ylabel("Cantidad")
plt.title("Entidades por Temas y Sector Tematic")

# Mostrar el gráfico
plt.show()
```
