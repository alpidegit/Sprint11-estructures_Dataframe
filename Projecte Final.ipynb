**Añadimos una columna que sea 1 si tienen presencia en redes y 0 si no**.

```python
# Utilizo .notna() para verificar si cada columna tiene un valor no nulo y el operador | = OR entre las condiciones de cada columna. 
# Convierto el resultado en 0 o 1 utilizando de las codniciones de las columnas .astype(int) y lo va asignando a la celda de la columna "Presencia_redes"
df_entidades["Presencia_redes"] = (df_entidades["XXSS_Instagram"].notna() | 
                                df_entidades["XXSS_Twitter"].notna() |
                                df_entidades["XXSS_Facebook"].notna() |
                                df_entidades["XXSS_Altres"].notna()).astype(int) 
df_entidades.head()
```
